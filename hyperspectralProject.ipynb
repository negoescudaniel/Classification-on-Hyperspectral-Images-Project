# Theoretical Introduction

**Hyperspectral images** are an advanced form of imaging used to capture detailed information about an object or scene by measuring the electromagnetic spectrum in hundreds of discrete spectral bands. This type of imaging provides a much higher level of detail compared to conventional RGB or multispectral images, which capture only a few spectral bands. Hyperspectral imaging analyzes light that is reflected, emitted, or transmitted across multiple wavelengths, from ultraviolet to infrared. Each pixel in a hyperspectral image contains a complete spectrum, which describes the object's response to various wavelengths.

## Examples of applications of hyperspectral images:

- **In agriculture**: *Monitoring plant health* (detecting water stress, diseases, or nutrient deficiencies); *Crop classification* (identifying crop types and estimating agricultural yield).

- **In environmental engineering**: *Pollution monitoring* (detecting contaminants in water, soil, or air); *Ecosystem mapping* (studying biodiversity and detecting ecological changes).

- **In the food industry**: *Quality control* (identifying contaminants and defects in food); *Food classification* (sorting foods based on composition and freshness).

- **In space exploration**: *Planetary surface analysis* (studying minerals and chemical composition on other planets); *Water detection* (identifying water and other resources on celestial bodies).

- **In medicine**: *Non-invasive diagnosis* (identifying skin abnormalities and affected tissues); *Guided surgery* (visualizing healthy versus affected tissue in real time).


# **Neural Networks**

Neural networks are computational models inspired by the human brain, widely used in machine learning for tasks such as image and speech recognition, natural language processing, and more.

The smallest component of a Neural Network (NN) is the neuron, which serves as a fundamental building block. A collection of neurons forms a layer of neurons. In practice, a neural network is built by stacking multiple layers with different configurations.

A basic NN consists of the following layers:

1. **Input Layer:** The layer that receives input data  
2. **Hidden Layers:** Layers between the input and output layers; their task is to process data received from the input layer  
3. **Output Layer:** Produces the final result  

## Neuron

- **Mathematical Definition**:
  - **Weighted Sum**: $ \sum_{i=1}^{n} w_i x_i + b $  
    - Where $ w_i $ are weights, $ x_i $ are inputs, and $ b $ is the bias.  
  - **Activation**: $ a = f(\sum_{i=1}^{n} w_i x_i + b) $  
    - Where $ f $ is the activation function.  

## Layer

- **Stacking Neurons**: A layer is a collection of neurons. In a fully connected layer, each neuron in one layer is connected to all neurons in the next layer.  
- **Mathematical Representation**:
  - For layer $ l $: $ a^l = f(W^l a^{l-1} + b^l) $  
    - Where $ W^l $ and $ b^l $ are the weights and biases for layer $ l $, and $ f $ is the activation function.  

## Activation Function

- **Purpose**: Introduces non-linearity, enabling the network to learn complex patterns.  
- **Types**:
  - **ReLU**: $ f(x) = \max(0, x) $  
  - **Sigmoid**: $ f(x) = \frac{1}{1 + e^{-x}} $  
  - **Softmax** (for the output layer in classification tasks): $ f(x_i)= \frac{e^{x_i}}{\sum_{j} e^{x_j}} $  

## Advantages

- **Ability to Learn Complex Patterns**: Neural networks are highly effective at capturing complex relationships between data.  
- **Versatility**: Can be used for a wide range of applications, including image recognition, text processing, and content generation.  
- **Scalability**: Can be scaled to work with large volumes of data, delivering impressive results on large datasets.  
- **End-to-End Learning**: Do not require manually defined features, as they can learn directly from raw data.  

## Disadvantages

- **High Computational Cost**: Training neural networks, especially deep ones, requires powerful hardware resources (GPUs).  
- **Prone to Overfitting**: If not well-managed, networks can memorize training data instead of generalizing.  
- **Requires Large Datasets**: The performance of neural networks depends on having a large and varied training dataset.  
- **Complexity**: Choosing the right architecture, activation functions, and parameters can be challenging and requires expertise.


# **Support Vector Machine (SVM)**

Support Vector Machine (SVM) is a machine learning algorithm used for classification and regression. It is known for its efficiency in separating both linear and non-linear data using the concept of an optimal hyperplane.

## Basic Concept

- **Hyperplane:** SVM creates a hyperplane that separates data into different classes with the largest possible **margin**.
- **Margin:** The distance between the closest data point (support vector) and the hyperplane. SVM maximizes this margin.

## Key Elements

### 1. **Decision Function**
SVM uses a decision function to classify data:
$$ f(x) = w^T x + b $$  
Where:  
- $ w $ is the weight vector.  
- $ b $ is the bias.  
- $ x $ represents the input data.

### 2. **Support Vectors**
- The data points closest to the hyperplane that influence its position and orientation.

### 3. **Kernel Trick**
For non-linear data, SVM uses kernel functions to transform the data into a higher-dimensional space where it can be linearly separated.  
Types of kernels:  
- **Linear:** $ K(x, y) = x^T y $  
- **Polynomial:** $ K(x, y) = (x^T y + c)^d $  
- **Radial Basis Function (RBF):** $ K(x, y) = e^{-\gamma \|x-y\|^2} $  

## Classification vs. Regression

1. **Classification:** SVM is used to separate data into distinct classes (e.g., image classification).
2. **Regression:** Extends the SVM concept for predicting continuous values.

## Advantages

- Efficient for large, sparse datasets.  
- Excellent performance in classification problems.  
- Flexible due to the use of kernels.

## Disadvantages

- High computational cost for large datasets.  
- Selecting the appropriate kernel can be challenging.


# **Indian Pines Hyperspectral Dataset**

The Indian Pines hyperspectral image dataset was downloaded from the following link: [Indian Pines Dataset](https://www.ehu.eus/ccwintco/index.php/Hyperspectral_Remote_Sensing_Scenes#Indian_Pines).

## Description

- **Dimensions:** 145 x 145 pixels.  
- **Spectral Bands:** 200 bands, each corresponding to a specific wavelength.  
- **Classes:** 16 different classes (e.g., wheat, forest, exposed soil, etc.).


###**Data Visualization**

```python
import scipy.io
import numpy as np
import matplotlib.pyplot as plt

# Read the .mat files
data = scipy.io.loadmat('Indian_pines_corrected.mat')['indian_pines_corrected']
labels = scipy.io.loadmat('Indian_pines_gt.mat')['indian_pines_gt']

# Dimensions and visualization
print(f"Image dimension: {data.shape}")  # (145, 145, 200)
print(f"Label dimension: {labels.shape}")  # (145, 145)

# Visualization of one spectral band
plt.imshow(data[:, :, 100], cmap='gray')  # Afișează banda spectrală 100
plt.title("Spectral band 100")
plt.show()

# Label Visualization
plt.imshow(labels, cmap='jet')
plt.title("Label (Ground Truth)")
plt.show()
```

###**Dimensionality reduction using the PCA(Principal Component Analysis) method**

```python
from sklearn.decomposition import PCA

# Parameters
num_components = 30  # Number of PCA components

# Reshape for PCA
# Transform the data from (145, 145, 200) to (145*145, 200)
reshaped_data = data.reshape(-1, data.shape[-1])

# Apply PCA
pca = PCA(n_components=num_components)
pca_data = pca.fit_transform(reshaped_data)  # Reduced dimenson

# Reshape back to the original dimension
pca_data_reshaped = pca_data.reshape(data.shape[0], data.shape[1], num_components)

# Rezulted dimensions
print(f"Dimension of the original data: {data.shape}")
print(f"Dimension after PCA: {pca_data_reshaped.shape}")  # (145, 145, num_components)

# Visualization of the fist principal components
plt.figure(figsize=(10, 6))
plt.imshow(pca_data_reshaped[:, :, 0], cmap='gray')  # First principal component
plt.title("First Principal Component (PCA)")
plt.colorbar()
plt.show()

# Varience for every component
explained_variance = pca.explained_variance_ratio_
print(f"Variance for the first {num_components} components: {sum(explained_variance) * 100:.2f}%")
```

###**Split the data into training set (85%) and testing set (15%)**

```python
from sklearn.model_selection import train_test_split

# Preparing the data and labels for splitting
# Flatten the 3D matrix into 2D for classification: (145*145, num_components)
X = pca_data_reshaped.reshape(-1, pca_data_reshaped.shape[-1])  # PCA data
y = labels.ravel()  # Labels

# Remove pixels without labels (class 0) to train only on valid data
valid_idx = y > 0
X = X[valid_idx]
y = y[valid_idx]

# Split the data into training set (85%) and testing set (15%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42, stratify=y)

# Check dimensions
print(f"Training set dimensions: {X_train.shape}, Labels: {y_train.shape}")
print(f"Testing set dimensions: {X_test.shape}, Labels: {y_test.shape}")
```

### **SVM Implementation and Accuracy Evaluation**

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import seaborn as sns

# 1. Create the SVM model
svm_model = SVC(kernel='rbf', C=1000, gamma='scale', random_state=42)

# 2. Train the model on the training set
svm_model.fit(X_train, y_train)

# 3. Make predictions on the testing set
y_pred = svm_model.predict(X_test)

# 4. Evaluate performance
# Confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)

# Overall accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Overall accuracy: {accuracy * 100:.2f}%")

# Detailed classification report
print("\nClassification report:\n")
print(classification_report(y_test, y_pred))

# 5. Visualize the confusion matrix
plt.figure(figsize=(10, 8))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=np.unique(y), yticklabels=np.unique(y))
plt.title("Confusion Matrix")
plt.xlabel("Predicted Classes")
plt.ylabel("Actual Classes")
plt.show()
```

### **Neural Network Model Implementation and Compilation**

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Ensure labels are in one-hot encoding format
y_train_fcnn = to_categorical(y_train - 1)  # One-hot encoding for training labels
y_test_fcnn = to_categorical(y_test - 1)  # One-hot encoding for testing labels

# Neural Network Architecture
model = Sequential()
model.add(Dense(500, activation='relu', input_shape=(X_train.shape[1],)))  # First dense layer
model.add(Dense(64, activation='relu'))  # Second dense layer
model.add(Dense(64, activation='relu'))  # Second dense layer
model.add(Dense(64, activation='relu'))  # Second dense layer
model.add(Dense(64, activation='relu'))  # Second dense layer
model.add(Dense(64, activation='relu'))  # Second dense layer
model.add(Dense(64, activation='relu'))  # Second dense layer
model.add(Dense(64, activation='relu'))  # Second dense layer
model.add(Dense(64, activation='relu'))  # Second dense layer
model.add(Dense(64, activation='relu'))  # Second dense layer
model.add(Dense(64, activation='relu'))  # Second dense layer
model.add(Dropout(0.2))  # Dropout to prevent overfitting
model.add(Dense(y_train_fcnn.shape[1], activation='softmax'))  # Output layer

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

###**Training of the Neural Network model**

```python
# Training of the NN model
history = model.fit(X_train, y_train_fcnn, validation_data=(X_test, y_test_fcnn), epochs=50, batch_size=64)
```

###**Evaluarea performanței metodei Neural Network**

```python
### **Evaluation of the Neural Network Method** # Evaluation

# Predictions on the test set
y_pred = model.predict(X_test)
y_pred_classes = tf.argmax(y_pred, axis=1).numpy()  # Choose the class with the highest probability
y_true_classes = tf.argmax(y_test_fcnn, axis=1).numpy()

# Confusion matrix and classification report
conf_matrix = confusion_matrix(y_true_classes, y_pred_classes)
print("\nClassification Report:\n")
print(classification_report(y_true_classes, y_pred_classes))

# Visualizing performance during training
plt.figure(figsize=(14, 6))

# Accuracy plot
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.legend()
plt.title("Accuracy Evolution")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")

# Loss plot
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.legend()
plt.title("Loss Evolution")
plt.xlabel("Epochs")
plt.ylabel("Loss")

plt.show()

# Final evaluation
test_loss, test_accuracy = model.evaluate(X_test, y_test_fcnn)
print(f'Test set accuracy: {test_accuracy * 100:.2f}%')
```

### **Comparison and Analysis**

From a performance perspective: SVM achieved ~1-2% better accuracy compared to the Neural Network. SVM is highly efficient for small to medium-sized datasets, where it finds the optimal separation margin between classes, especially with the RBF kernel, which effectively models non-linear decision boundaries. Neural Networks (NN) require larger datasets to excel. While NN can model complex relationships, smaller datasets can lead to overfitting or underfitting, affecting performance.

From a development time perspective, Neural Networks require numerous experiments to select the optimal number of layers, neurons, learning rate, dropout, etc. This process can be time and resource-intensive, whereas SVM has only a few key hyperparameters to tune (e.g., C, gamma). Training NN is significantly slower than SVM, especially for larger architectures or when using limited GPUs. SVM has an advantage in such scenarios, particularly for moderately sized data.

#### **How Methods Were Affected**

**SVM:**

- **Advantages:**
  - Dimensionality reduction (PCA) improved performance by eliminating spectral redundancy. SVM excels when data is reduced to a subset of relevant features.
  - The RBF kernel effectively utilizes spectral information to find optimal separation boundaries between complex classes.
  - SVM works well on preprocessed datasets with reduced dimensionality, which was the case here.

- **Limitations:**
  - SVM classifies each pixel individually without considering spatial relationships between pixels, which can lead to errors in noisy or homogeneous regions.

**Neural Networks (NN):**

- **Advantages:**
  - Neural Networks can learn complex relationships from data, including non-linear interactions between spectral bands.
  - With sufficient data, NN can learn more advanced patterns than SVM.

- **Limitations:**
  - While PCA reduced data dimensionality, NN still required multiple adjustments to optimize performance in such cases.
  - The available classification data (only validated pixels) was relatively limited.

#### **Conclusions**

**SVM:** Was more suitable for this preprocessed dataset, efficiently utilizing the spectral features extracted via PCA.

**NN:** Was more sensitive to the limited amount of data and required architectural adjustments to exploit complex information effectively.
